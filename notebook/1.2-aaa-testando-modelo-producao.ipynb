```python
# importando bibliotecas

# dados
import pandas as pd
import numpy as np

# pre-processamento
from sklearn.preprocessing import Normalizer

# algoritmos cluster
from sklearn.cluster import KMeans

# métricas de validação
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import calinski_harabasz_score

# Visualizações
import matplotlib.pyplot as plt
import seaborn as sns

# salvar e carregar modelo

import pickle
```

```python
df = pd.read_csv('../data/CC GENERAL.csv')
df.head()
```

```python
# remover colunas

df.drop(columns=['CUST_ID','TENURE'], inplace=True)
df.head()
```

```python
# substituindo dados faltantes

df.fillna(df.median(), inplace=True)
df.isna().sum()
```

```python
# normalizando os dados
df_normalize = Normalizer().fit_transform(df.values)
print(df_normalize)
```

```python
model_pkl_file = "../model/finalized_model.pkl" 

# carregando modelo pickle
with open(model_pkl_file, 'rb') as file:  
    model = pickle.load(file)
```

```python
model
```

```python
# aplicando o predict

y_pred = model.fit_predict(df_normalize)
```

```python
labels = model.labels_
silhouette_score(df_normalize,labels, metric='euclidean') # métric é o método de calculo da distância entre os pontos
```

```python
davies_bouldin_score(df_normalize, labels)
```

```python
calinski_harabasz_score(df_normalize, labels)
```

```python
df['cluster'] = labels
```
